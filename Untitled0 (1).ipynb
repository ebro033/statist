# №1

#В атмосфере планеты Синелябор иногда случаются торнадо. Длительные
#наблюдения показывают, что обычно происходит 4.5 торнадо в одну
#синеляборскую неделю.
#На Синелябор упал крупный метеорит, и астрономы предполагают, что это
#повлияло на атмосферные явления, вследствие чего частота торнадо на
#планете изменилась. Они наблюдали за атмосферой Синелябора две
#недели после падения метеорита и хотят сделать вывод о влиянии
#метеорита на торнадо.

# a)

1. Нулевая Гипотеза (H0): Падение метеорита не повлияло на частоту торнадо на планете Синелябор. Средняя частота торнадо осталась 4.5 в неделю.

2. Альтернативная Гипотеза (H1): Падение метеорита повлияло на частоту торнадо на планете Синелябор. Средняя частота торнадо не равна 4.5 в неделю.

# b)

```python
#z-тест


import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Исходные данные
mu = 4.5 * 2  # Среднее количество торнадо за две недели
sigma = 2 * np.sqrt(2)  # Стандартное отклонение за две недели
alpha = 0.05  # Уровень статистической значимости

# Критические значения
z_critical_lower = norm.ppf(alpha / 2)
z_critical_upper = norm.ppf(1 - alpha / 2)
tornadoes_critical_lower = mu + z_critical_lower * sigma
tornadoes_critical_upper = mu + z_critical_upper * sigma

# Генерация значений для оси x (количество торнадо)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Плотность вероятности для оси y
y = norm.pdf(x, mu, sigma)

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Нормальное распределение')
plt.fill_between(x, y, where=(x < tornadoes_critical_lower) | (x > tornadoes_critical_upper), color='red', alpha=0.5, label='Критические области')
plt.axvline(x=tornadoes_critical_lower, color='red', linestyle='--', label=f'Нижняя граница = {tornadoes_critical_lower:.2f}')
plt.axvline(x=tornadoes_critical_upper, color='red', linestyle='--', label=f'Верхняя граница = {tornadoes_critical_upper:.2f}')
plt.title("Распределение количества торнадо за две недели на Синеляборе")
plt.xlabel("Количество торнадо")
plt.ylabel("Плотность вероятности")
plt.legend()
plt.show()
```

#c)

```python
# Наблюдаемое количество торнадо за две недели
observed_tornadoes = 14

# Z-статистика для наблюдаемого значения
z_observed = (observed_tornadoes - mu) / sigma

# P-значение (двусторонний тест)
p_value_observed = 2 * norm.cdf(-abs(z_observed))
p_value_observed
```

```python
# Расчет p-value
# Z-статистика
z_observed_14 = (14 - mu) / sigma

# P-значение (двусторонний тест)
p_value_observed_14 = 2 * norm.cdf(-abs(z_observed_14))

p_value_observed_14
```

# №2

#При открытии сети кафе «Енот-поласкун», где можно гладить енотов,
#владельцы хотят проверить гипотезу, что в среднем посетитель успевает
#погладить 8 енотов за время одного сеанса: если это значение больше, то
#надо будет заводить дополнительных енотов. Первые несколько
#посетителей успели погладить следующее количество енотов за один
#сеанс: 10, 11, 6, 8, 13, 10, 7, 9, 10, 12, 8

#a)

1. Нулевая Гипотеза : Среднее количество енотов, которых гладят посетители за один сеанс, равно 8. То есть \( \mu_0 = 8 \).

2. Альтернативная Гипотеза : Среднее количество енотов, которых гладят посетители за один сеанс, больше 8.


#b)

```python
import numpy as np
from scipy.stats import t

# Данные о количестве енотов, которых успели погладить посетители
data = np.array([10, 11, 6, 8, 13, 10, 7, 9, 10, 12, 8])

# Среднее выборочное
mean_sample = np.mean(data)

# Стандартное отклонение выборки с поправкой Бесселя
std_sample = np.std(data, ddof=1)

# Размер выборки
n = len(data)

# Предполагаемое среднее значение в нулевой гипотезе
mu_0 = 8

# Расчет t-статистики
t_statistic = (mean_sample - mu_0) / (std_sample / np.sqrt(n))
t_statistic
```

#c)

```python
# Расчет критического значения t-статистики для одностороннего теста при уровне значимости 5%

t_critical = t.ppf(1 - 0.05, df=n-1)  # df - степени свободы
t_critical
```

#d)

```python
# Используем данные из предыдущего примера для расчета t-статистики

# Расчет среднего выборочного и стандартного отклонения выборки
mean_sample = data.mean()
std_sample = data.std(ddof=1)

# Размер выборки
n = len(data)

# Предполагаемое среднее значение в нулевой гипотезе
mu_0 = 8

# Расчет t-статистики
t_statistic = (mean_sample - mu_0) / (std_sample / np.sqrt(n))

# Критическое значение t-статистики для одностороннего теста при уровне значимости 5%
t_critical = t.ppf(1 - 0.05, df=n-1)

# Вывод о нулевой гипотезе
reject_null = t_statistic > t_critical

t_statistic, t_critical, reject_null
```

```python
import matplotlib.pyplot as plt

# Диапазон значений для оси x
x = np.linspace(-4, 4, 100)
# t-распределение с n-1 степенями свободы
y = t.pdf(x, df=n-1)

# Визуализация
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='t-распределение')
plt.fill_between(x, y, where=(x > t_critical), color='red', alpha=0.5, label='Критическая область (отклоняем H0)')
plt.axvline(x=t_statistic, color='green', linestyle='-', label=f't-статистика = {t_statistic:.2f}')
plt.axvline(x=t_critical, color='red', linestyle='--', label=f'Критическое значение = {t_critical:.2f}')
plt.title("Визуализация t-теста для кафе 'Енот-поласкун'")
plt.xlabel("Значения t-статистики")
plt.ylabel("Плотность вероятности")
plt.legend()
plt.show()
```

#e)

```python
from scipy.stats import ttest_1samp

# Проведение одновыборочного t-теста
t_test_result = ttest_1samp(data, mu_0)

# Результаты теста
t_statistic_scipy = t_test_result.statistic
p_value_scipy = t_test_result.pvalue

t_statistic_scipy, p_value_scipy
```
